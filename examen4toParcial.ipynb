```python
print("ejercicio 1")
import numpy as np
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn import metrics
visitas=[50,38,28,8,8,16,5,2]
valor=[134844,55526,5523,4067,611,4459,2195,1225]

n=len(visitas)
n1=len(valor)
visitas=np.array(visitas)
valor=np.array(valor)

#sumando los vectores de x y y
sumvisitas=sum(visitas)
sum_valor=sum(valor)

#obteniendo los promedios
promvisitas=sumvisitas/n
prom_valor=sum_valor/n1

parte_abajo=sum(pow(visitas-promvisitas,2))

parte_arriba=sum((visitas-promvisitas)*(valor-prom_valor))
b=(parte_arriba)/(parte_abajo)

#print("b = ",b)
a=prom_valor-b*promvisitas
#print("a = ",a)
print("a y b")
print("a= ",a)
print("b= ",b)

print("\n cuando una empresa tiene 10 visitas su ganancia en dolares es",(a+b*10),"\n")
print("a) Diagrama de dispersion")
plt.plot(visitas,valor,'o',Label='datos')
plt.plot(visitas,a+b*visitas,Label='Ajuste')
plt.xlabel('visitas')
plt.ylabel('valor')
plt.grid()#para poner al grafico lineas
plt.legend()
plt.show()

aa=a+b*50
aa1=a+b*51
aa2=a+b*52
aa3=a+b*53

print("diferencias",(aa-aa1))
print("diferencias",(aa1-aa2))

print("diferencias",(aa2-aa3))

print("se puede ver que cuando hay visitas cada dia pues el dinero en dolares va aumentando cada dia en",(aa1-aa2)," dolares")
```

```python

print("ejercicio 2")
visitas=[50,38,28,8,8,16,5,2]
valor=[134844,55526,5523,4067,611,4459,2195,1225]
visitas=np.array(visitas)
valor=np.array(valor)
x=np.array(list(zip(visitas,valor))).reshape(len(visitas),2)
plt.plot(visitas,valor,'o',Label='datos')
#plt.plot(x1,a+b*x1,Label='Ajuste')
plt.xlabel('visitas')
plt.ylabel('valor')
plt.grid()#para poner al grafico lineas
plt.legend()
plt.show()
#algoritmo kmeans
k=3
modelo_kmeans=KMeans(n_clusters=k).fit(x)
for i,l in enumerate(modelo_kmeans.labels_):
    print("(x1,x2)->:clase")
    print("{0},{1}->:{2}".format(visitas[i],valor[i],l))
predicir=modelo_kmeans.predict(x)
print("\ngrupos o cluster que se tiene\n",predicir,"\n")
prueba=[[3,5]]
prediccion=modelo_kmeans.predict(prueba)
print("prediccion:({0},{1})".format(prueba,prediccion),"\n")
#calculando centroides
print("\n")
print(modelo_kmeans.cluster_centers_)

centro=modelo_kmeans.cluster_centers_
#tarea graficar los puntos centroides
print("\ngrafica de lo centroides mas los puntos")
plt.plot(visitas,valor,'o',Label='datos')
ce=centro[0]
plt.plot(ce[0],ce[1],marker='o', mfc='red',label="centroides")    

for i in range(0,len(centro)):
    cxy=centro[i]
    plt.plot(cxy[0],cxy[1],marker='o', mfc='red')    
#plt.plot(x1,a+b*x1,Label='Ajuste')
plt.xlabel('visitas')
plt.ylabel('valor')
plt.grid()#para poner al grafico lineas
plt.legend()
plt.show()
```
